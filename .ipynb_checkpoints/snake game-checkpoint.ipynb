```python
import pygame
import time
import random

def our_snake(snake_block, snake_list):
    for x in snake_list:
        pygame.draw.rect(dis, black, [x[0], x[1], snake_block, snake_block])

def message(msg, color):
    font_style = pygame.font.SysFont(None, 50)
    mesg = font_style.render(msg, True, color)
    dis.blit(mesg, [dis_width/2 -350, dis_height/2 -25])
    
def your_score(score):
    score_font = pygame.font.SysFont("comicsansms", 35)
    value = score_font.render("Your Score: " + str(score), True, blue)
    dis.blit(value, [10, 0])
    
def high_score(score):
    score_font = pygame.font.SysFont("comicsansms", 35)
    value = score_font.render("High Score: " + str(score), True, green)
    dis.blit(value, [560, 0])

def gameLoop():
    game_over = False
    game_close = False
    
    snake_list = []
    
    x1 = dis_width / 2
    y1 = dis_height / 2
    
    snake_list.append([x1, y1])
    len_snake = 1
    
    foodx = round(random.randrange(0, dis_width-snake_block)/10)*10
    foody = round(random.randrange(0, dis_height-snake_block)/10)*10
    
    last_key = pygame.K_RIGHT
    global high
    
    while not game_over:
        
        while game_close == True:
#             dis.fill(white)
            message("You Lost! Press Q to Quit or C to Play Again", red)
            your_score(len_snake - 1)
            high_score(high)
            pygame.display.update()
            
            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key ==  pygame.K_q:
                        game_over = True
                        game_close = False
                    elif event.key == pygame.K_c:
                        gameLoop()
                if event.type == pygame.QUIT:
                    game_over = True
                    game_close = False
         
        x1_change = 0
        y1_change = 0
            
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True

            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT and last_key != pygame.K_RIGHT:
                    x1_change = -snake_block
                elif event.key == pygame.K_RIGHT and last_key != pygame.K_LEFT:
                    x1_change = snake_block
                elif event.key == pygame.K_DOWN and last_key != pygame.K_UP:
                    y1_change = snake_block
                elif event.key == pygame.K_UP and last_key != pygame.K_DOWN:
                    y1_change = -snake_block
                    
                dis.fill(white)
            
                x1 += x1_change
                y1 += y1_change
                
                if x1_change!=0 or y1_change!=0:
                    snake_list.pop(0)
                    if [x1, y1] in snake_list:
                        game_close = True
                
                    snake_list.append([x1, y1])
                    last_key = event.key
        
  
                
        if x1_change==0 and y1_change == 0:
            if last_key == pygame.K_LEFT:
                x1_change = -snake_block
            elif last_key == pygame.K_RIGHT:
                x1_change = snake_block
            elif last_key == pygame.K_DOWN:
                y1_change = snake_block
            elif last_key == pygame.K_UP:
                y1_change = -snake_block
            
            dis.fill(white)
            
            x1 += x1_change
            y1 += y1_change
                
            snake_list.pop(0)
                
            if [x1, y1] in snake_list:
                game_close = True
                
            snake_list.append([x1, y1])
    

        if x1 >= dis_width or x1 <=0 or y1>=dis_height or y1<=0:
            game_close = True
                
      

            
        if x1==foodx and y1==foody:
#             print("Yummy!")
            foodx = round(random.randrange(0, dis_width-snake_block)/10)*10
            foody = round(random.randrange(0, dis_height-snake_block)/10)*10
            snake_list.append([x1, y1])
            len_snake += 1
            if len_snake > high:
                high = len_snake-1
                
            
        our_snake(snake_block, snake_list)
        pygame.draw.rect(dis, red, [foodx, foody, snake_block, snake_block])
        your_score(len_snake - 1)
        high_score(high)
        pygame.display.update()

        clock.tick(snake_speed)
    pygame.quit()            

    

if __name__ == "__main__":
    pygame.init()
    
    dis_width = 800
    dis_height = 600

    white = (255, 255, 255)
    red = (255, 0, 0)
    blue = (0, 0, 205)
    black = (0, 0, 0)
    green = (0, 205, 0)


    dis = pygame.display.set_mode((dis_width, dis_height))
    dis.fill(white)
    pygame.display.update()
    pygame.display.set_caption("Snake game")

    clock = pygame.time.Clock()
    snake_speed = 20
    snake_block = 10
    high = 0
    gameLoop()
    

```
